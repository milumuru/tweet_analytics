```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#CSV読み込み

df = pd.read_csv("today.csv", index_col=False)
```

```python
df.describe()
```

```python
#favが多い順に並べる
df = df.sort_values(by='fav', ascending=False)
```

```python
# 上位100件を見てみるよ！
df.head(100)
```

```python
#いいねが339(平均)以上の対象取得
df1=df[df['fav'] >339]

df1.to_csv("fav_339.csv")
```

```python
df1.tail(100)
```

```python
# いいねされてシェアされるコンテンツを散布図で可視化する

df.plot(kind='scatter', x='RT', y='fav', xlim=[0,30000], ylim=[340,104000], title="Favorites and Retweets")
```

```python
#RT:2000、いいね：7000(分かりにくいため、明らかに多い数字は分かりにくいため、除外)
df1.plot(kind='scatter', x='RT', y='fav', xlim=[0,2000], ylim=[340,7000], title="Favorites and Retweets")
```

```python
#RT:1000、いいね：7000(分かりにくいため、明らかに多い数字は分かりにくいため、除外)
df1.plot(kind='scatter', x='RT', y='fav', xlim=[0,1000], ylim=[340,7000], title="Favorites and Retweets")
```

```python
#RT:2000、いいね：2500(分かりにくいため、明らかに多い数字は分かりにくいため、除外)
df1.plot(kind='scatter', x='RT', y='fav', xlim=[100,2000], ylim=[340,2500], title="Favorites and Retweets")
```

```python
df1['rate'] = 100*df1['RT']/df1['fav']
```

```python
df1.head()
```

```python
df1.head(100)
```

```python
# rate率が高い順に並べる
df1= df1.sort_values(by=["rate"], ascending=False)
```

```python
df1.head(100)
```

```python
# 少ない順に並べる
df2= df1.sort_values(by=["rate"], ascending=True)
df2.head(100)
```

```python
# 見えてきたのでブログにまとめるためにCSVを保存する

df1.to_csv("highRTrate.csv", encoding='utf_8_sig')
df2.to_csv("lowRTrate.csv", encoding='utf_8_sig')
```
